```python
import cv2 as cv
from PIL import Image,ImageFilter
import matplotlib.pyplot as plt
import numpy as np


high=Image.open("C:\\Users\\Saket\\Desktop\\assignemnt 1\\nikola.jpeg")
high=high.convert("L")
plt.subplot(3,3,1)
plt.imshow(high,cmap="gray")
plt.title("image 1")
plt.axis("off")


low=Image.open("C:\\Users\\Saket\\Desktop\\assignemnt 1\\oppen.jpeg")
low=low.convert("L")

plt.subplot(3,3,4)
plt.imshow(low,cmap="gray")
plt.title("image 2")
plt.axis("off")


high_conv = high.filter( ImageFilter.Kernel( (3,3), (-1,-1,-1,-1,8,-1,-1,-1,-1), 1, 0) )
plt.subplot(3,3,3)
plt.imshow(high_conv, cmap="gray")
plt.title("High Pass")
plt.axis("off")


low_conv = low.filter( ImageFilter.Kernel( (5,5), (1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9), 1, 0) )
plt.subplot(3,3,6)
plt.imshow(low_conv, cmap="gray")
plt.title("Low Pass")
plt.axis("off")


low_conv = low_conv.resize(high_conv.size)




high_np = np.array(high_conv, dtype=np.float32)
low_np = np.array(low_conv, dtype=np.float32)

alpha = 0.06
beta=0.3

combined = alpha * high_np + beta * low_np
combined = np.clip(combined, 0, 255).astype(np.uint8)
combined_image = Image.fromarray(combined)

plt.subplot(3,3,8)
plt.imshow(combined_image, cmap="gray")
plt.title("Combined")
plt.axis("off")

plt.show()




```
