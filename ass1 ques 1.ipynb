```python
import matplotlib.pyplot as plt
import os
from PIL import Image

def show(name,n,m,i,Title):
    plt.subplot(n,m,i)
    plt.imshow(name,cmap="gray")
    plt.title(Title)
    plt.axis("off")




p=[]
img_title=[]
x=1
path="C:\\Users\\Saket\\Desktop\\assignemnt 1\\pokemon_images"
for i in os.listdir(path):
    img_path=os.path.join(path,i)
    p.append(img_path)
    img_title.append("image"+str(x))
    x=x+1


for name , i , Title in zip(p,range(1,5),img_title):

    img = Image.open(name)
    img = img.convert("L")

    show(img,2,2,i,Title)
```
